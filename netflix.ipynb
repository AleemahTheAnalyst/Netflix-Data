````python
#Import all necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
````

<p>Creating a list of years, dating from 2011 - 2020 and an average random movie length. Thereafter, the years and movie lengths are compiled into a dictionary named <code>'movie_dict'</code> and then converted into a dataframe, <code>'durations_df'</code> and then displayed.</p>

````python
#Create the years and durations lists
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019,2020]
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]

#Create a dictionary with the two lists
movie_dict = pd.DataFrame({
    'years' : years, 'durations': durations,
},
    columns = ['years', 'durations'])

#Print the dictionary
movie_dict
````

````python
#Create a dataframe from the dictionary
durations_df = movie_dict

#Print the DataFrame
durations_df
````

### Visualization of the duration_dataframe

<p>This shows the matplotlib library with <code>'Years'</code> on the x-axis and <code>'Durations'</code> on the y-axis, titled <code>"Netflix Movie Durations 2011 - 2020"</code>.</p>

````python
 #Import matplotlib.pyplot under its usual alias and create a figure
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(11,7))

# Draw a line plot of release_years and durations
plt.plot(years, durations, color='purple', marker='o')

# Create a title
plt.title('Release Years Vs Durations')
plt.xlabel(' Years ')

plt.ylabel(' Durations ')

# Show the plot
plt.show
````

````python
# Subset the DataFrame for type "Movie"
netflix_df_movies_only = netflix_df[(netflix_df['type'] == 'Movie')]
netflix_df_movies_only.head()


# Select only these columns of interest (title, country, genre, release_year and duration)
netflix_movies_col_subset= netflix_df[['title', 'country', 'genre', 'release_year', 'duration']]

# Print the first five rows of the new DataFrame
netflix_movies_col_subset.to_csv('netflix_movies_col_subset', index = False)
netflix_movies_col_subset.head()
````

#### Create a Scatterplot

Let us visualise the <code>"realease_year"</code> on the x-axis vs <code>"movie duration"</code> on the y-axis, titled<code> "Movie Duration by Year of Realease"</code> with a scatter plot using matplotlib visualisation module.

````python
# Create a figure and increase the figure size
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus year
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], color='cyan', marker='x')
plt.xlabel('release_year')
plt.ylabel('duration')

# Create a title
plt.title('Release Year vs Duration')

# Show the plot
plt.show
````

<p>Subset netflix_movies_col_subset create a new dataframe; short_movies, for movies with a duration of less than 60 mins. The first 20 rows printed, to get an overview of the movie types with short duration.</p>

````python
# Filter for durations shorter than 60 minutes
short_movies = netflix_movies_col_subset[netflix_movies_col_subset['duration']<60]

# Print the first 20 rows of short_movies
short_movies.head(20)
````

#### Marking Non-feature films

<p>Append Colors to genre by using a <code>for</code> loop to iterate through the rows and conditional statments to append the colors per genre based on 'Children', <code>'Documentaries',</code> <code>'Stand-Up'</code>, <code>'Others'</code>.</p>

````python
# Define an empty list
colors = []

# Children, Stand-Up, Documentaries
# Iterate over rows of netflix_movies_col_subset
for row, ser in netflix_movies_col_subset.iterrows() :
    if ser['genre'] == 'Children':
        colors.append('red')
    elif ser['genre'] == 'Documentaries':
        colors.append('blue')
    elif ser['genre'] == 'Stand-Up':
        colors.append('green')
    else:
        colors.append('black')
        
# Inspect the first 10 values in your list        
print(colors[0:10])
````

````python
# Set the figure style and initalize a new figure
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus release_year
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], color=colors, marker='o')
plt.xlabel('release_year')
plt.ylabel('duration')

# Create a title and axis labels
plt.title('Release_Year Vs Duration')

# Show the plot
plt.show
````
